## Adding Arr_Delay_Tag Column

```python

df_1_flights['Arr_Delay_Tag'] = df_1_flights['Arr_Delay'].apply(lambda x: 1 if x > 0 else 0)
```

## Adding Season Column

```python
df_1_flights['FlightDate'] = pd.to_datetime(df_1_flights['FlightDate'])

# Extract the month from the 'FlightDate' column
df_1_flights['Month'] = df_1_flights['FlightDate'].dt.month

# Function to determine the season based on the month
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'


df_1_flights['Season'] = df_1_flights['Month'].apply(get_season)
```

# Airline Analysis

## Number of Delayed Flights per Airline - On Departure

```python
import matplotlib.pyplot as plt
#Checking only delayed flights with label 1
delayed_flights = df_1_flights[df_1_flights['Dep_Delay_Tag'] == 1]

# Count the number of delayed flights per airline
delay_counts_per_airline = delayed_flights['Airline'].value_counts().sort_values(ascending=False)

# Horizontal bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(delay_counts_per_airline.index, delay_counts_per_airline.values, color='skyblue')

# Adding the count of delayed flights
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.0f}', 
             va='center', ha='left')

plt.xlabel('Number of Delayed Flights')
plt.ylabel('Airline')
plt.title('Number of Delayed Flights per Airline - On Departure')
plt.gca().invert_yaxis()  # To have the biggest values at the top
plt.show()

```

## Total Flights and Percentage of Delayed Flights per Airline

```python
# Total flights and delayed flights per airline
total_flights_per_airline = df_1_flights['Airline'].value_counts().sort_index()
delayed_flights_per_airline = df_1_flights[df_1_flights['Dep_Delay_Tag'] == 1]['Airline'].value_counts().sort_index()

# Percentage of delayed flights per airline
percentage_delayed = (delayed_flights_per_airline / total_flights_per_airline) * 100

# Combine into a DataFrame for sorting
df_plot = pd.DataFrame({
    'Total_Flights': total_flights_per_airline,
    'Delayed_Flights': delayed_flights_per_airline.reindex(total_flights_per_airline.index, fill_value=0),
    'Percentage_Delayed': percentage_delayed
})

# Sort by percentage of delayed flights in ascending order
df_plot = df_plot.sort_values(by='Percentage_Delayed')

# Plotting Prep
airlines = df_plot.index
total_flights = df_plot['Total_Flights'].values
delayed_flights = df_plot['Delayed_Flights'].values
percentage_delayed = df_plot['Percentage_Delayed'].values

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))

bars1 = plt.barh(airlines, total_flights, color='skyblue', label='Total Flights')
bars2 = plt.barh(airlines, delayed_flights, color='lightcoral', label='Delayed Flights')

# Adding the percentage of delayed flights to the end of the red part of each bar
for i, bar in enumerate(bars2):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{percentage_delayed[i]:.1f}%', va='center', ha='left')

plt.xlabel('Number of Flights')
plt.ylabel('Airline')
plt.title('Total Flights and Percentage of Delayed Flights per Airline')
plt.legend(loc='upper right')
plt.gca().invert_yaxis()  # biggest percentage on top
plt.show()

```

## Merging Flight Dataset and Weather Dataset

```python
df_weather = pd.read_csv ('weather_meteo_by_airport.csv')
```

```python

df_weather['time'] = pd.to_datetime(df_weather['time']).dt.date

# Merge datasets on date and airport_id
df_1_flights['FlightDate_only'] = df_1_flights['FlightDate'].dt.date
df_merged = pd.merge(df_1_flights, df_weather, left_on=['FlightDate_only', 'Dep_Airport'], right_on=['time', 'airport_id'])
```

```python
# Drop useless merging columns
df_merged = df_merged.drop(columns=['airport_id', 'FlightDate_only', 'time'])
```

## Reorder Columns

```python
new_order = [
    'FlightDate',
    'Month',
    'Day_Of_Week',
    'Season',
    'Airline',
    'Dep_Airport',
    'Dep_CityName',
    'DepTime_label',
    'Dep_Delay',
    'Dep_Delay_Tag',
    'Dep_Delay_Type',
    'Arr_Airport',
    'Arr_CityName',
    'Arr_Delay',
    'Arr_Delay_Tag',
    'Arr_Delay_Type',
    'Flight_Duration',
    'Distance_type',
    'Delay_Carrier',
    'Delay_Weather',
    'Delay_NAS',
    'Delay_Security',
    'Delay_LastAircraft',
    'Tail_Number',
    'Manufacturer',
    'Model',
    'Aicraft_age',
    'tavg',
    'tmin',
    'tmax',
    'prcp',
    'snow',
    'wdir',
    'wspd',
    'pres'
]

# Reorder the columns
df_merged = df_merged[new_order]
```

## Correlation Matrix

```python
# Just delayed flights
filtered_df = df_merged[df_merged['Dep_Delay_Tag'] == 1]

# Selecting relevant columns for the correlation matrix
columns_of_interest = ['Dep_Delay', 'Arr_Delay', 'Flight_Duration', 'Delay_Carrier', 'Delay_Weather', 'Delay_NAS', 'Delay_Security', 'Delay_LastAircraft', 'tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wdir', 'wspd', 'pres']

# Compute the correlation matrix
correlation_matrix = filtered_df[columns_of_interest].corr()

# Plotting the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix for Delayed Flights')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cross-tabulation of 'Dep_Delay_Type' and 'Distance_type'
delay_distance_crosstab = pd.crosstab(df_merged['Distance_type'], df_merged['Dep_Delay_Type'])

# Normalize the cross-tabulation to get the percentage distribution
delay_distance_crosstab_normalized = delay_distance_crosstab.div(delay_distance_crosstab.sum(axis=1), axis=0) * 100

# Stacked bar chart
ax = delay_distance_crosstab_normalized.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Paired')

# Adding labels and title
plt.xlabel('Distance Type')
plt.ylabel('Percentage')
plt.title('Correlation between Departure Delay Type and Distance Type')
plt.legend(title='Departure Delay Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

# Display the percentage values
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f%%', label_type='center')

plt.tight_layout()
plt.show()
```
